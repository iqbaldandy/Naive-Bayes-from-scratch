```
#probability acc condition
acc = df[df['condition']=='acc']
acc_p = acc.shape
acc_p = acc_p[0]
acc_predicted = acc_p/700

#probability unacc condition
unacc = df[df['condition']=='unacc']
unacc_p = unacc.shape
unacc_p = unacc_p[0]
unacc_predicted = unacc_p/700
```

```
answer = []
a = 0
b = 0
c = 0
d = 0
for i,j in dftest.iterrows():
  predict_a_l = []
  predict_u_l = []
  if j['buying'] == 'vhigh':
    predict_a_l.append(bvhigh_a_c/acc_p)
    predict_u_l.append(bvhigh_u_c/unacc_p)
  elif j['buying'] == 'high':
    predict_a_l.append(bhigh_a_c/acc_p)
    predict_u_l.append(bhigh_u_c/unacc_p)
  elif j['buying'] == 'med':
    predict_a_l.append(bmed_a_c/acc_p)
    predict_u_l.append(bmed_u_c/unacc_p)
  elif j['buying'] == 'low':
    predict_a_l.append(blow_a_c/acc_p)
    predict_u_l.append(blow_u_c/unacc_p)
  if j['maintenance'] == 'vhigh':
    predict_a_l.append(mvhigh_a_c/acc_p)
    predict_u_l.append(mvhigh_u_c/unacc_p)
  elif j['maintenance'] == 'high':
    predict_a_l.append(mhigh_a_c/acc_p)
    predict_u_l.append(mhigh_u_c/unacc_p)
  elif j['maintenance'] == 'med':
    predict_a_l.append(mvhigh_a_c/acc_p)
    predict_u_l.append(mvhigh_u_c/unacc_p)
  elif j['maintenance'] == 'low':
    predict_a_l.append(mlow_a_c/acc_p)
    predict_u_l.append(mlow_u_c/unacc_p)
  if j['doors'] == '2':
    predict_a_l.append(d2_a_c/acc_p)
    predict_u_l.append(d2_u_c/unacc_p)
  elif j['doors'] == '3':
    predict_a_l.append(d3_a_c/acc_p)
    predict_u_l.append(d3_u_c/unacc_p)
  elif j['doors'] == '4':
    predict_a_l.append(d4_a_c/acc_p)
    predict_u_l.append(d4_u_c/unacc_p)
  elif j['doors'] == '5more':
    predict_a_l.append(d5more_a_c/acc_p)
    predict_u_l.append(d5more_u_c/unacc_p)
  if j['persons'] == '2':
    predict_a_l.append(p2_a_c/acc_p)
    predict_u_l.append(p2_u_c/unacc_p)
  elif j['persons'] == '4':
    predict_a_l.append(d4_a_c/acc_p)
    predict_u_l.append(d4_u_c/unacc_p)
  elif j['persons'] == 'more':
    predict_a_l.append(pmore_a_c/acc_p)
    predict_u_l.append(pmore_u_c/unacc_p)
  if j['luggage'] == 'small':
    predict_a_l.append(lsmall_a_c/acc_p)
    predict_u_l.append(lsmall_u_c/unacc_p)
  elif j['luggage'] == 'med':
    predict_a_l.append(lmed_a_c/acc_p)
    predict_u_l.append(lmed_u_c/unacc_p)
  elif j['luggage'] == 'big':
    predict_a_l.append(lbig_a_c/acc_p)
    predict_u_l.append(lbig_u_c/unacc_p)
  if j['safety'] == 'low':
    predict_a_l.append(slow_a_c/acc_p)
    predict_u_l.append(slow_u_c/unacc_p)
  elif j['safety'] == 'med':
    predict_a_l.append(smed_a_c/acc_p)
    predict_u_l.append(smed_u_c/unacc_p)
  elif j['safety'] == 'high':
    predict_a_l.append(shigh_a_c/acc_p)
    predict_u_l.append(shigh_u_c/unacc_p)
  resulta1 = np.prod(predict_a_l)
  resultu1 = np.prod(predict_u_l)
  resulta2 = resulta1*acc_predicted
  resultu2 = resultu1*unacc_predicted
  if resulta2 > resultu2:
    hasil = 'acc'
    answer.append(hasil)
  elif resulta2 < resultu1:
    hasil = 'unacc'
    answer.append(hasil)
  if j['condition']=='acc' and hasil == 'acc':
    a += 1
  elif j['condition'] == 'acc' and hasil == 'unacc':
    b += 1
  elif j['condition'] == 'unacc' and hasil == 'acc':
    c += 1
  elif j['condition'] == 'unacc' and hasil == 'unacc':
    d += 1
```

```
print(answer)
```

```
accuracy = (a + d)/(a + b + c + d)
precision = a/(a+c)
recall = a/(a+b)
f_measure = (2*recall*precision)/(recall + precision)

print("hasil")
print("akurasi : ",accuracy)
print("precision : ",precision)
print("recall : ",recall)
print("f_measure: ",f_measure)
```
